```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y = load_diabetes(return_X_y = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
r2_score(y_test, reg.predict(X_test))
```

```python
# buliding my own SGD Regressor
class SGDRegressor:
  def __init__(self, ephocs = 100, learning_rate = 0.01):
    self.coef_ = None
    self.intercept_ = None
    self.lr = learning_rate
    self.ephocs = ephocs

  def fit(self, X_train, y_train):
    # assuming intercept to " 0 "
    self.intercept_ = 0

    # creating an array having number of coef_ = cols in X_train
    self.coef_ = np.ones(X_train.shape[1])

    for i in range(self.ephocs):

      for j in range(X_train.shape[0]):
                # selecting a random row
                idx = np.random.randint(0,X_train.shape[0])

                # predicting for the random row
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_

                intercept_der = -2 * (y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)

    print(self.coef_, self.intercept_)

  def predict(self, X_test):
    return np.dot(X_test, self.coef_) + self.intercept_
```

```python
sgd = SGDRegressor(learning_rate=0.01, ephocs=50)
```

```python
sgd.fit(X_train, y_train)
```

```python
r2_score(y_test, sgd.predict(X_test))
```
